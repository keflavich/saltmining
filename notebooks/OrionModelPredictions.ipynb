```python
# June 8, 2022 version - discussion w/Ceccarelli about optically thin dust disk
plotwl = np.logspace(0, 5, 1000)*u.um
plotfreq = plotwl.to(u.GHz, u.spectral())
rovib_range_plot = (plotwl>25*u.um) & (plotwl<45*u.um)

tbldata = []

for density in (1e6, 1e7, 1e8,1e9,1e10,1e11)*u.cm**-3:
    for column in (1e12,1e13,1e14,1e15)*u.cm**-2:
        for disktem in (100, 200, 300, 500, 1000, 2000, 3000, 4000)*u.K:
            for tgas in (50, 70, 100, 150,)*u.K:
                tmod = disktem
                beta=2
                plotbg = bgfunc_mod(plotfreq, disktem=disktem, beta=beta)
                disk,star,cmb = bgfunc_mod(plotfreq, disktem=disktem, beta=beta, return_components=True)
                rr.background_brightness = bgfunc_mod(freq, disktem=disktem, beta=beta)
                rslt = (rr(density=density, column=column, temperature=tgas, tbg=None))



                fig7 = pl.figure(7, figsize=(10,12))
                fig7.clf()
                ax1 = pl.subplot(3,1,1)
                pl.loglog(plotwl, plotbg, linestyle='-', label='Model radiation field')
                axlims = ax1.axis()
                pl.loglog(plotwl, disk, linestyle='-', label='Disk')
                pl.loglog(plotwl, star, linestyle='-', label='Star')
                pl.loglog(plotwl, cmb, linestyle='-', label='CMB')
                ax1.axis(axlims)

                pl.loglog(wl, rr.background_brightness, marker='.', linestyle='none')
                pl.xlabel("Wavelength [$\mu$m]")
                pl.ylabel("Background Brightness\n[{}]".format(rr.background_brightness.unit.to_string(format='latex')))
                pl.legend(loc='best')
                ax2 = pl.subplot(3,1,2)

                # introduce an arbitrary scaling value to obtain 'reasonable' column densities
                scaling = 1e12 * u.cm**-2

                pl.plot(rslt['upperstateenergy'].data, np.log10(rslt['upperlevelpop'].data * scaling.value), '.', label=None, alpha=0.15)

                #nacl = Vamdc.query_molecule('NaCl$')
                frqs, einsteinAij, degeneracies, EU, partfunc = get_molecular_parameters('NaCl, v=0-15', catalog='CDMS', fmin=85*u.GHz, fmax=360*u.GHz)

                fitcolrot, fittrot,_,_ = fit_tex(u.Quantity(rslt[vone & obs]['upperstateenergy'], u.K), rslt[vone &
                                                                                    obs]['upperlevelpop']*scaling,
                        partition_func=partfunc,plot=True, min_nupper=1e-50, labeltemonly=True)
                fitcolvib, fittvib,_,_ = fit_tex(u.Quantity(rslt[Jeight]['upperstateenergy'], u.K),
                        rslt[Jeight]['upperlevelpop']*scaling, partition_func=partfunc,plot=True,
                        min_nupper=1e-50, color='k', labeltemonly=True)
                pl.legend(loc='upper right')

                pl.xlabel("E$_U$ [K]")
                pl.ylabel("log upper state population")
                pl.ylim(6,11)

                ax3 = pl.subplot(3,1,3)
                pl.plot(rslt['upperstateenergy'].data, (rslt['T_B']), '.', label=None, alpha=0.15)
                pl.plot(rslt['upperstateenergy'].data, (rslt['Tex'] * (1-np.exp(-rslt['tau']))), '.', label=None, alpha=0.15)
                pl.xlabel("E$_U$ [K]")
                pl.ylabel("$T_B$ [K]")

                tbldata.append({'density': density,
                                'column': column,
                                'tgas': tgas,
                                'disktem': disktem,
                                'tvib': fittvib,
                                'trot': fittrot,
                                'colvib': fitcolvib,
                                'colrot': fitcolrot,
                               })
                #print(tbldata)
                tbl = Table(tbldata)
                tbl.write("OrionDiskLVGModelPredictionsModifiedBB.ecsv", overwrite=True)

                pl.tight_layout()
                fig7.savefig(paths.fpath(f'simulated_populations_with_modbb_radiation_field_tmod{int(tmod.value):d}_tgas{int(tgas.value):d}_column{column.value:0.1e}_density{density.value:0.1e}.pdf'), bbox_inches='tight')
                fig7.savefig(paths.fpath(f'simulated_populations_with_modbb_radiation_field_tmod{int(tmod.value):d}_tgas{int(tgas.value):d}_column{column.value:0.1e}_density{density.value:0.1e}.svg'), bbox_inches='tight')
                fig7.savefig(paths.fpath(f'simulated_populations_with_modbb_radiation_field_tmod{int(tmod.value):d}_tgas{int(tgas.value):d}_column{column.value:0.1e}_density{density.value:0.1e}.png'), bbox_inches='tight')
```
